# 用于测试该设备torch是否运行良好的代码

PyTorch官网：https://pytorch.org/get-started/locally/

```python
import torch
import sys
import time
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
device
```

```python
# 检查GPU是否可用
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('使用的设备:', device)

# 如果设备是GPU，打印GPU的名称
if device.type == 'cuda':
    print('GPU名称:', torch.cuda.get_device_name(0))

# 创建一个大的随机张量用于测试
size = 10000  # 矩阵大小为10000x10000
a = torch.randn(size, size)

# 在CPU上进行矩阵乘法并计时
start_time = time.time()
a_cpu = a.matmul(a)
cpu_time = time.time() - start_time
print(f"CPU计算时间: {cpu_time:.6f}秒")

# 如果GPU可用，则在GPU上进行同样的操作
if device.type == 'cuda':
    a_gpu = a.to(device)  # 将张量移动到GPU
    start_time = time.time()
    a_gpu.matmul(a_gpu)
    torch.cuda.synchronize()  # 等待cuda上的操作完成，确保计时准确
    gpu_time = time.time() - start_time
    print(f"GPU计算时间: {gpu_time:.6f}秒")
    print(f"GPU加速比: {cpu_time/gpu_time:.2f}倍")
else:
    print("没有可用的GPU设备进行加速计算。")
```
