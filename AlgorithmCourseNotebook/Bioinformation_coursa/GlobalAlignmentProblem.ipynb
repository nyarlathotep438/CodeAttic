```python
import sys
sys.setrecursionlimit(10000)

def longest_common_subsequence(s1, s2):
    m, n = len(s1), len(s2)
    # 初始化DP表
    dp = [[0] * (n + 1) for _ in range(m + 1)]  
    
    # 填充DP表
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    
    # 回溯构造LCS
    lcs = []
    i, j = m, n
    while i > 0 and j > 0:
        if s1[i-1] == s2[j-1]:
            lcs.append(s1[i-1])
            i -= 1
            j -= 1
        elif dp[i-1][j] > dp[i][j-1]:
            i -= 1
        else:
            j -= 1
    
    return ''.join(reversed(lcs))

# 测试验证
print(longest_common_subsequence("GAGA", "GAT"))
```

```python
def global_alignment(match_reward, mismatch_penalty, indel_penalty, s1, s2):
    m, n = len(s1), len(s2)
    # 正确初始化DP表和回溯路径表
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    trace = [[0] * (n + 1) for _ in range(m + 1)]  # 0=对角线, 1=向上, 2=向左
    
    # 初始化第一列（全为indel罚分）
    for i in range(1, m + 1):
        dp[i][0] = dp[i-1][0] - indel_penalty
        trace[i][0] = 1  # 向上回溯
    
    # 初始化第一行（全为indel罚分）
    for j in range(1, n + 1):
        dp[0][j] = dp[0][j-1] - indel_penalty
        trace[0][j] = 2  # 向左回溯
    
    # 填充DP表
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            # 计算三种情况的得分
            if s1[i-1] == s2[j-1]:
                match_score = dp[i-1][j-1] + match_reward
            else:
                match_score = dp[i-1][j-1] - mismatch_penalty
            
            delete_score = dp[i-1][j] - indel_penalty  # 在s1中插入gap
            insert_score = dp[i][j-1] - indel_penalty   # 在s2中插入gap
            
            # 取最大值并记录路径
            max_score = max(match_score, delete_score, insert_score)
            dp[i][j] = max_score
            
            # 记录回溯方向
            if max_score == match_score:
                trace[i][j] = 0
            elif max_score == delete_score:
                trace[i][j] = 1
            else:
                trace[i][j] = 2
    
    # 回溯构建比对结果
    align1, align2 = [], []
    i, j = m, n
    while i > 0 or j > 0:
        if i > 0 and j > 0 and trace[i][j] == 0:  # 对角线
            align1.append(s1[i-1])
            align2.append(s2[j-1])
            i -= 1
            j -= 1
        elif i > 0 and (trace[i][j] == 1 or j == 0):  # 向上
            align1.append(s1[i-1])
            align2.append('-')
            i -= 1
        else:  # 向左
            align1.append('-')
            align2.append(s2[j-1])
            j -= 1
    
    # 反转比对结果
    align1 = ''.join(reversed(align1))
    align2 = ''.join(reversed(align2))
    
    return dp[m][n], align1, align2

# 示例使用
if __name__ == "__main__":
    # 匹配奖励=1, 错配惩罚=1, 插入缺失惩罚=2
    score, s1_align, s2_align = global_alignment(1, 1, 2, "GAGA", "GAT")
    
    print(score)
    print(s1_align)
    print(s2_align)
```

```python
def local_alignment(s1, s2):
    INDEL_PENALTY = 5
    
    m, n = len(s1), len(s2)
    # 初始化动态规划表和回溯表
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    trace = [[0] * (n + 1) for _ in range(m + 1)]
    
    # 填充动态规划表
    max_score = -1
    max_i, max_j = 0, 0
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            # 计算三种情况的得分
            match_score = dp[i - 1][j - 1] + PAM250[s1[i - 1]][s2[j - 1]]
            delete_score = dp[i - 1][j] - INDEL_PENALTY
            insert_score = dp[i][j - 1] - INDEL_PENALTY
            
            # 确定最大值并更新
            scores = [0, match_score, delete_score, insert_score]
            max_val = max(scores)
            dp[i][j] = max_val
            
            # 记录最大值位置（用于回溯）
            if max_val > max_score:
                max_score = max_val
                max_i, max_j = i, j
            
            # 记录转移方向
            if max_val == 0:
                trace[i][j] = 0
            elif max_val == match_score:
                trace[i][j] = 1
            elif max_val == delete_score:
                trace[i][j] = 2
            else:  # insert_score
                trace[i][j] = 3
    
    # 回溯构建比对结果
    align1, align2 = [], []
    i, j = max_i, max_j
    while i > 0 or j > 0:
        if trace[i][j] == 1:  # 对角线（匹配）
            align1.append(s1[i - 1])
            align2.append(s2[j - 1])
            i -= 1
            j -= 1
        elif trace[i][j] == 2:  # 向上（s1字符对应空位）
            align1.append(s1[i - 1])
            align2.append('-')
            i -= 1
        elif trace[i][j] == 3:  # 向左（s2字符对应空位）
            align1.append('-')
            align2.append(s2[j - 1])
            j -= 1
        else:  # trace[i][j] == 0（停止回溯）
            break
    
    # 反转得到正确顺序
    align1 = ''.join(reversed(align1))
    align2 = ''.join(reversed(align2))
    return max_score, align1, align2

# 示例使用
if __name__ == "__main__":
    s1 = "MEANLY"
    s2 = "PENALTY"
    score, align1, align2 = local_alignment(s1, s2)
    
    print(f"最大得分: {score}")
    print(f"比对结果 1: {align1}")
    print(f"比对结果 2: {align2}")
```

```python
def edit_distance(s1, s2):
    m, n = len(s1), len(s2)
    # 正确的初始化：创建 (m+1) x (n+1) 的二维列表
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # 初始化边界条件
    for i in range(m + 1):
        dp[i][0] = i  # s2为空字符串时，删除s1的所有字符
    for j in range(n + 1):
        dp[0][j] = j  # s1为空字符串时，插入s2的所有字符
    
    # 动态规划填表
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                # 字符相同，无需操作
                dp[i][j] = dp[i - 1][j - 1]
            else:
                # 取三种操作的最小值
                dp[i][j] = min(
                    dp[i - 1][j] + 1,      # 删除操作
                    dp[i][j - 1] + 1,      # 插入操作
                    dp[i - 1][j - 1] + 1   # 替换操作
                )
    
    return dp[m][n]

# 测试示例
print(edit_distance("kitten", "sitting"))  # 输出: 3 (k→s, e→i, +g)
```

```python
def overlap_alignment(match, mismatch, indel, v, w):
    n, m = len(v), len(w)
    # Initialize DP table
    dp = [[0] * (m + 1) for _ in range(n + 1)]
    backtrack = [[None] * (m + 1) for _ in range(n + 1)]

    # Fill DP table
    for i in range(1, n + 1):
        dp[i][0] = -i * indel
        backtrack[i][0] = "up"
    for j in range(1, m + 1):
        dp[0][j] = -j * indel
        backtrack[0][j] = "left"

    for i in range(1, n + 1):
        for j in range(1, m + 1):
            score_sub = dp[i-1][j-1] + (match if v[i-1] == w[j-1] else -mismatch)
            score_del = dp[i-1][j] - indel
            score_ins = dp[i][j-1] - indel
            dp[i][j] = max(score_sub, score_del, score_ins)

            if dp[i][j] == score_sub:
                backtrack[i][j] = "diag"
            elif dp[i][j] == score_del:
                backtrack[i][j] = "up"
            else:
                backtrack[i][j] = "left"

    # Find best overlap: suffix of v (last row) vs prefix of w (last col)
    max_score = float("-inf")
    max_pos = (0, 0)

    # last row
    for j in range(m + 1):
        if dp[n][j] > max_score:
            max_score = dp[n][j]
            max_pos = (n, j)

    # last column
    for i in range(n + 1):
        if dp[i][m] > max_score:
            max_score = dp[i][m]
            max_pos = (i, m)

    # Traceback
    i, j = max_pos
    v_aln, w_aln = "", ""
    while i > 0 and j > 0:
        if backtrack[i][j] == "diag":
            v_aln = v[i-1] + v_aln
            w_aln = w[j-1] + w_aln
            i -= 1
            j -= 1
        elif backtrack[i][j] == "up":
            v_aln = v[i-1] + v_aln
            w_aln = "-" + w_aln
            i -= 1
        elif backtrack[i][j] == "left":
            v_aln = "-" + v_aln
            w_aln = w[j-1] + w_aln
            j -= 1
        else:
            break

    return max_score, v_aln, w_aln


# --- Test with Sample Input ---
match, mismatch, indel = 1, 1, 2
v = "GAGA"
w = "GAT"

score, v_aln, w_aln = overlap_alignment(match, mismatch, indel, v, w)
print(score)
print(v_aln)
print(w_aln)
```

```python
def overlap_alignment(v, w, match_reward, mismatch_penalty, indel_penalty):
    len_v = len(v)
    len_w = len(w)
    
    # 初始化DP表和回溯表
    dp = [[float('-inf')] * (len_w + 1) for _ in range(len_v + 1)]
    backtrack = [[''] * (len_w + 1) for _ in range(len_v + 1)]
    
    # 初始化第一行（i=0）：v开头可跳过，得分为0
    for j in range(len_w + 1):
        dp[0][j] = 0
        backtrack[0][j] = 'skip_v'
    
    # 填充DP表
    for i in range(1, len_v + 1):
        for j in range(1, len_w + 1):
            # 考虑匹配/替换
            if v[i-1] == w[j-1]:
                score_diag = dp[i-1][j-1] + match_reward
            else:
                score_diag = dp[i-1][j-1] + mismatch_penalty
            
            # 考虑在v中插入空格（跳过w[j-1]）
            score_left = dp[i][j-1] + indel_penalty
            # 考虑在w中插入空格（跳过v[i-1]）
            score_up = dp[i-1][j] + indel_penalty
            
            # 选择得分最高的转移
            if score_diag >= score_left and score_diag >= score_up:
                dp[i][j] = score_diag
                backtrack[i][j] = 'diag'
            elif score_left >= score_up:
                dp[i][j] = score_left
                backtrack[i][j] = 'left'
            else:
                dp[i][j] = score_up
                backtrack[i][j] = 'up'
    
    # 在最后一行（i=len_v）找到最大得分
    max_score = max(dp[len_v])
    j_end = dp[len_v].index(max_score)
    
    # 回溯构建对齐结果
    i, j = len_v, j_end
    align_v, align_w = [], []
    
    while j > 0:  # 回溯至w的开头（j=0）
        if backtrack[i][j] == 'diag':
            align_v.append(v[i-1])
            align_w.append(w[j-1])
            i -= 1
            j -= 1
        elif backtrack[i][j] == 'up':
            align_v.append(v[i-1])
            align_w.append('-')
            i -= 1
        else:  # 'left'
            align_v.append('-')
            align_w.append(w[j-1])
            j -= 1
    
    # 反转对齐结果（回溯是逆序）
    align_v.reverse()
    align_w.reverse()
    
    # 转换为字符串
    align_v_str = ''.join(align_v)
    align_w_str = ''.join(align_w)
    
    return max_score, align_v_str, align_w_str

# 示例
if __name__ == "__main__":
    # 输入参数
    match = 1
    mismatch = -1
    indel = -2
    v = "GAGA"
    w = "GAT"
    
    # 计算最优重叠对齐
    score, align_v, align_w = overlap_alignment(v, w, match, mismatch, indel)
    
    # 输出结果
    print(f"Optimal Score: {score}")
    print(f"Aligned v suffix: {align_v}")
    print(f"Aligned w prefix: {align_w}")
```

```python
def fitting_alignment(v, w, match, mismatch, indel):
    n, m = len(v), len(w)
    dp = [[0] * (m+1) for _ in range(n+1)]
    
    # 初始化
    for i in range(1, n+1):
        dp[i][0] = i * indel
    
    # 填充DP表
    for i in range(1, n+1):
        for j in range(1, m+1):
            score_match = dp[i-1][j-1] + (match if v[i-1] == w[j-1] else mismatch)
            score_indel_v = dp[i-1][j] + indel  # 在v中插入空格（跳过w的字符）
            score_indel_w = dp[i][j-1] + indel  # 在w中插入空格（跳过v的字符）
            dp[i][j] = max(score_match, score_indel_v, score_indel_w)
    
    # 找到最大得分（v的最后一行的最大值）
    max_score = max(dp[n])
    return max_score

# 参数设置
match = 1
mismatch = -1
indel = -1
v = "ACGTCG"
w = "GTTGGATTACGAATCGATATCTGTTTG"

# 计算最大得分
max_score = fitting_alignment(v, w, match, mismatch, indel)
print("最大拟合对齐得分:", max_score)  # 输出: 0
```
