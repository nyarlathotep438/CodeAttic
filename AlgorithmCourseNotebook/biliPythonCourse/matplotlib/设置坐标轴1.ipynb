```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
```

```python
plt.figure()
plt.plot(x,y2)
plt.plot(x, y1 ,color = 'red', linewidth = 1.0, linestyle = '--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.xlabel('I am x')
plt.ylabel('I am y')
plt.show()
```
