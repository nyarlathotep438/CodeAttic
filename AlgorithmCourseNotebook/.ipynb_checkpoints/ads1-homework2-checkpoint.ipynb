```python
def readGenome(filename):
    genome = ''
    with open(filename, 'r') as f:
        for line in f:
            # ignore header line with genome information
            if not line[0] == '>':
                genome += line.rstrip()
    return genome
```

```python
def naive_with_counts(p, t):
    occurrences = []
    num_alignments = 0
    num_character_comparisons = 0
    
    for i in range(len(t) - len(p) + 1):
        num_alignments += 1
        match = True
        for j in range(len(p)):
            num_character_comparisons += 1
            if t[i+j] != p[j]:
                match = False
                break
        if match:
            occurrences.append(i)
    
    return (occurrences, num_alignments, num_character_comparisons)
```

```python
def boyer_moore_with_counts(p, p_bm, t):
    occurrences = []
    num_alignments = 0
    num_character_comparisons = 0
    
    i = 0
    while i <= len(t) - len(p):
        num_alignments += 1
        shift = 1
        mismatched = False
        
        # 从右向左比较
        for j in range(len(p)-1, -1, -1):
            num_character_comparisons += 1
            if p[j] != t[i+j]:
                # mismatch，计算跳跃
                bc_shift = p_bm.bad_character_rule(j, t[i+j])
                gs_shift = p_bm.good_suffix_rule(j)
                shift = max(shift, bc_shift, gs_shift)
                mismatched = True
                break
        
        if not mismatched:
            # 完全匹配
            occurrences.append(i)
            shift = p_bm.match_skip()
        
        i += shift
    
    return (occurrences, num_alignments, num_character_comparisons)
```

```python
from kmer_index import Index
```

```python
def query_index_2mm(p, t, index):
    k = index.k
    hits = set()
    for offset in [0, 8, 16]:
        kmer = p[offset:offset+k]
        for pos in index.query(kmer):
            start = pos - offset
            if start < 0 or start + len(p) > len(t):
                continue
            mismatches = 0
            for i in range(len(p)):
                if p[i] != t[start+i]:
                    mismatches += 1
                    if mismatches > 2:
                        break
            if mismatches <= 2:
                hits.add(start)
    return hits
```

```python
pattern = 'GGCGCGGTGGCTCACGCCTGTAATCCCAGCACTTTGGGAGGCCGAGG'
index = Index(chr1_genome, 8)
hits = query_index_2mm(pattern, chr1_genome, index)
print(len(hits))
```

```python
index = Index(chr1_genome, 8)
p = 'GGCGCGGTGGCTCACGCCTGTAAT'

total_hits = 0
for offset in [0, 8, 16]:
    kmer = p[offset:offset+8]
    hits = index.query(kmer)
    total_hits += len(hits)

print(total_hits)
```

```python
import bisect
   
class SubseqIndex(object):
    """ Holds a subsequence index for a text T """
    
    def __init__(self, t, k, ival):
        """ Create index from all subsequences consisting of k characters
            spaced ival positions apart.  E.g., SubseqIndex("ATAT", 2, 2)
            extracts ("AA", 0) and ("TT", 1). """
        self.k = k  # num characters per subsequence extracted
        self.ival = ival  # space between them; 1=adjacent, 2=every other, etc
        self.index = []
        self.span = 1 + ival * (k - 1)
        for i in range(len(t) - self.span + 1):  # for each subseq
            self.index.append((t[i:i+self.span:ival], i))  # add (subseq, offset)
        self.index.sort()  # alphabetize by subseq
    
    def query(self, p):
        """ Return index hits for first subseq of p """
        subseq = p[:self.span:self.ival]  # query with first subseq
        i = bisect.bisect_left(self.index, (subseq, -1))  # binary search
        hits = []
        while i < len(self.index):  # collect matching index entries
            if self.index[i][0] != subseq:
                break
            hits.append(self.index[i][1])
            i += 1
        return hits
```

```python
def query_subseq(p, t, subseq_ind):
    hits = []
    ival = subseq_ind.ival

    # 对 subseq 的多个 offset 进行查询
    for offset in range(ival):
        # 传入从 offset 开始的 pattern 片段
        hits.extend(subseq_ind.query(p[offset:]))

    return hits, len(hits)
```
