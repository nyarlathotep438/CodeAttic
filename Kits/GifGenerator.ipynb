```python
import imageio.v2 as imageio
import os

folder = r"C:\PhD Project\MAIN PROCSSS\VolcanoPlots"

filenames = sorted([
    os.path.join(folder, f)
    for f in os.listdir(folder)
    if f.endswith(".png")
])

frames = [imageio.imread(f) for f in reversed(filenames)]

outfile = f"{folder}/volcano_animation_1_0.5.gif"
imageio.mimsave(outfile, frames, fps=2)
```

```python
import imageio.v2 as imageio
import os

folder = r"C:\PhD Project\MAIN PROCSSS\VolcanoPlots"

filenames = sorted([
    os.path.join(folder, f)
    for f in os.listdir(folder)
    if f.endswith(".png")
])

frames = [imageio.imread(f) for f in filenames]

outfile = os.path.join(folder, "volcano_animation.mp4")
imageio.mimsave(outfile, frames, fps=1.5, codec="libx264")
```
