```python
import matplotlib.pyplot as plt  
import numpy as np  
  
# 数据  
data = [3, 0, 5, -2, -4, 1, 3, 1, -8, 3, -8]  
  
# 计算四分位数和IQR  
Q1 = np.percentile(data, 25)  
Q3 = np.percentile(data, 75)  
IQR = Q3 - Q1  
  
# 计算疑似异常值的界限  
lower_limit = Q1 - 1.5 * IQR  
upper_limit = Q3 + 1.5 * IQR  
  
# 找出疑似异常值  
outliers = [(i, x) for i, x in enumerate(data) if x < lower_limit or x > upper_limit]  
  
# 创建箱型图  
fig, ax = plt.subplots()  
  
# 箱型图默认是垂直的，通过设置vert=True使其水平，并旋转标签  
ax.boxplot(data, vert=True, showfliers=False, patch_artist=True)  
  
# 填充箱体颜色（可选）  
colors = ['#3498db', '#95a5a6', '#e74c3c', '#2ecc71', '#555']  
for patch, color in zip(ax.patches, colors):  
    patch.set_facecolor(color)  
  
# 旋转x轴标签（可选）  
for tick in ax.get_xticklabels():  
    tick.set_rotation(90)  
  
# 标注异常值  
for i, outlier in outliers:  
    ax.text(i, outlier, f'{outlier}', va='center', ha='right',  
            bbox=dict(boxstyle='round', facecolor='white', edgecolor='0.3'))  
  
# 设置标题和标签  
plt.title('Boxplot with Labeled Outliers')  
plt.xlabel('Data')  
plt.ylabel('Values')  
  
# 调整图形布局，确保异常值标签可见  
plt.tight_layout()  
  
# 显示图形  
plt.show()
```
