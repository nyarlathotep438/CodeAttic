```python
# class for a dog object
class Dog:
    
    # initialisation method with internal data
    def __init__(self, petname, temp) :
        self.name = petname;
        self.temperature = temp;
        
    # get status
    def status(self):
        print("dog name is",self.name)
        print("dog temperture is",self.temperature)
        pass
    
    #set temperture
    def setTemperture(self,temp):
        self.temperature = temp;
        pass
    
    # dogs can bark()
    def bark(self):
        print("woof!")
        pass
    
    pass
    
```

```python
# create a new dog object from the Dog class
lassie = Dog("Lassie",37)
```

```python
lassie.status()
```

```python
lassie.setTemperture(40)
```

```python
lassie.status()
```

```python
lassie.bark()
```
