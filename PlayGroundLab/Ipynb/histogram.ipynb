```python
import matplotlib.pyplot as plt  
  
ranges = [(40, 60), (60, 70), (70, 80), (80, 85), (85, 90)]  # 范围  
frequencies = [100, 60, 90, 70, 60]  # 对应每个范围的频率
  
# 提取范围的边界用于绘图  
left_edges = [r[0] for r in ranges]  
widths = [r[1] - r[0] for r in ranges]  # 每个条形的宽度  
  
# 绘制条形图  
plt.bar(left_edges, frequencies, width=widths, edgecolor='black', align='edge')  
  
# 添加标题和标签  
plt.title('Frequency Histogram')  
plt.xlabel('Range')  
plt.ylabel('Frequency')  
  
# 设置x轴刻度为范围的中点  
mid_points = [(r[0] + r[1]) / 2 for r in ranges]  
plt.xticks(mid_points, [f'{l}-{h}' for l, h in ranges])  
  
# 显示图形  
plt.show()
```
