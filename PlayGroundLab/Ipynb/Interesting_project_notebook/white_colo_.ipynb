# 📌 为什么是这个比例？
为了让你理解背后的逻辑，我也把比例结构解释一下：

| 材料 | 作用 | 常见比例 | 你这份配方 |
|------|------|-----------|-------------|
| 奶油芝士 | 稳定结构、厚度 | 1 份 | 180 g |
| 白巧克力 | 甜味 + 冷藏后凝固 | 0.8–1 份 | 150 g |
| 淡奶油 | 空气感、轻盈 | 1.3–1.5 份 | 300 ml |

这个比例保证：  
- **不需要明胶也能稳定**  
- **不会太甜**（白巧克本身甜度高）  
- **入口即化，不会像重芝士蛋糕那样厚重**
