# ERNIE-Image 测试

```python
prompt="一张日式平涂动漫插画，展现一个绿色眼睛、留着深蓝色齐整短发，身穿蓝色连体泳衣的高挑少女。她微笑着在蔚蓝的高空中以游泳的动作飞行。"
```

```python
import torch
from diffusers import ErnieImagePipeline
import logging
logging.disable(logging.WARNING)

pipe = ErnieImagePipeline.from_pretrained(
    "Baidu/ERNIE-Image",
    torch_dtype=torch.bfloat16
).to("cuda")  
```

```python
prompt="""A full-body anime character illustration of Rimi Ushigome from BanG Dream!, standing in a gentle, shy pose. She holds a small glass jar of strawberry jam with both hands, lifting it softly to her chest as if presenting it with pride. Her expression is warm, slightly embarrassed, with soft pink blush on her cheeks and sparkling eyes.

The outfit is an original design inspired by British dessert culture and afternoon tea aesthetics. A pastel cream and strawberry-pink color palette. A layered dress resembling a Victorian sponge cake: soft cream-colored bodice, strawberry-red accents, delicate lace trims like whipped cream, and skirt layers shaped like cake tiers. Subtle strawberry motifs embroidered along the hem. A small apron inspired by tea-time tablecloth patterns, with scalloped lace edges. Accessories include a tiny silver spoon hairpin, a ribbon shaped like a jam label, and a miniature tag tied to her waist like a tea-room ticket.

Her hairstyle follows Rimi’s canon look: short, soft, slightly curled dark hair with pink highlights, matching her gentle personality. Lighting is soft and warm, like afternoon tea sunlight through a café window. Background is minimal, a faint pastel gradient to highlight the outfit design.

Full-body, clean lineart, high-resolution character sheet style, centered composition, no extreme perspective, outfit details clearly visible.
"""
```

```python
image = pipe(
    prompt= prompt,
    height=2048,
    width=1024,
    num_inference_steps=30,
    guidance_scale=4.0,
    use_pe=True,
).images[0]

image.save("test_Rimi.png")
```

# ERNIE-Image-Turbo 测试

```python
import os
import torch
from diffusers import ErnieImagePipeline
import logging
logging.disable(logging.WARNING)

os.environ["HF_HOME"] = "C:/Users/Admin/ernie_image_Turbo_cache"

pipe = ErnieImagePipeline.from_pretrained(
    "Baidu/ERNIE-Image-Turbo",
    torch_dtype=torch.bfloat16,
).to("cuda")
```

```python
prompt="""一幅日系动漫风格插画。
背景为跆拳道馆，
画面中一个穿连体竞速泳衣的纤细少女跪在地上。少女们双手背在身后，低着头弯着腰，表情不安又害羞。大量穿白色跆拳道服的初中小女孩在少女身后站成一排，得意地比着剪刀手。
画面整体呈现出精细的平涂上色（Cell Shading）风格，线条干净利落。"""
```

```python
save_dir = "batch_test_11 "
os.makedirs(save_dir, exist_ok=True)

# 批量生成 10 张图
for i in range(10):
    img = pipe(
        prompt=prompt,
        height=1024,
        width=1024,
        num_inference_steps=8,
        guidance_scale=1.0,
        use_pe=True
    ).images[0]
    
    # 文件名：img_01.png ~ img_10.png
    filename = f"img_{i+1:02d}.png"
    img.save(os.path.join(save_dir, filename))

print("批量生成完成！")
```
