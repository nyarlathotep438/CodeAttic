```python
import torch
from diffusers import ErnieImagePipeline
import warnings

warnings.filterwarnings(
    "ignore",
    message="Expected types for pe",
)

warnings.filterwarnings(
    "ignore",
    message="Expected types for text_encoder",
)


pipe = ErnieImagePipeline.from_pretrained(
    "Baidu/ERNIE-Image",
    torch_dtype=torch.bfloat16
)

pipe.to("cuda")   

image = pipe(
    prompt="请画一幅日式平涂动漫插画：一个绿色眼睛、留着深蓝色齐整短发，身穿蓝色连体泳衣和绿色飞行夹克，背着战斗机机翼的高挑少女，表情得意地在万里无云的高空中以游泳的动作飞行。",
    height=1024,
    width=1024,
    num_inference_steps=30,
    guidance_scale=4.0,
    use_pe=False,
).images[0]

image.save("test5.png")
```
