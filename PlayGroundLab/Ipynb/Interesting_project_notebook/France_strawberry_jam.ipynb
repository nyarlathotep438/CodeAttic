 这份食谱可以做出约 2.5 罐果酱(果酱瓶的容量180ml）

✎ 材料/ Ingredients
草莓 500g (去蒂后的净重)
水 50g
细砂糖 400g
柠檬汁 25g

✎  
1. 把草莓倒在一个大盆里，在流动的水下，用手轻轻滑动，清洗每一颗草莓，也要把蒂头冲一下水，由于草莓不耐水，切勿长时间泡水
2. 把草莓沥干，并用厨房纸巾吸干水份，手法要轻柔，尽可能不要压坏果肉
3. 用小刀去掉蒂头，并集合在一个大盆里
4. 加入细砂糖、柠檬汁和水，充份拌匀
5. 把事先好的圆形烘焙纸直接贴合草莓覆盖上，再包上保鲜膜，送进冰箱糖渍一整天
6. 第二天，把糖渍一天了的草莓倒入锅中，以中小火煮到微滚，离火，盖上烘焙纸及保鲜膜，放凉后移到冰箱里冰藏一天
7. 第三天，在熬煮果酱前先来消毒 瓶子，把果酱瓶跟盖子放入煮沸的热水中，至少煮个2-3分钟，然后取出放在干净的布或纸巾上晾干
8. 接着用滤网把渍草莓的糖浆过滤出来，先用中大火把糖浆煮滚. 煮滚后续煮到糖浆明显变得浓稠些，温度会来到105°C，由于草莓水份较少，熬煮后的稠度会比其它水果来得稀一点
9. 加入草莓果肉，以中小火再次煮到滚，煮沸后再续煮五分钟，这时可以滴几滴果酱在盘子或汤勺上，放凉后用手指在果酱上画条线，如果线条分明就是胶化完成了，关火
10. 煮好的果酱要尽快装瓶、盖上盖子并旋紧，然后把果酱瓶倒扣、放凉，放凉后可以送进冰箱保鲜

# 🍓 400g 草莓 + lime 版本法式草莓果酱（改编配方）

## 🧂 **材料（Ingredients）**

- **草莓 400g**（去蒂后净重）  
- **水 40g**  
- **细砂糖 320g**  
- **lime（酸橙）汁 20g**  
  - 若你的 lime 不够，可以用 **15g**，也完全可行  
  - 酸度足够，不影响凝固

---

# 🔥 **制作步骤（改编自页面三天法式做法）**

## **Day 1：糖渍草莓**
1. 草莓洗净、沥干，用厨房纸轻轻吸干水分（避免压坏果肉）。  
2. 去蒂后放入大盆。  
3. 加入 **320g 糖 + 20g lime 汁 + 40g 水**，轻轻拌匀。  
4. 贴上一张圆形烘焙纸，再盖保鲜膜。  
5. 冷藏 **24 小时**。

---

## **Day 2：第一次加热**
1. 将腌好的草莓连同糖浆倒入锅中。  
2. 中小火加热至 **微滚**（小气泡）。  
3. 离火，重新盖上烘焙纸与保鲜膜。  
4. 放凉后再冷藏 **24 小时**。

---

## **Day 3：正式熬煮 + 装瓶**
### 🔪 1. 消毒瓶子  
把果酱瓶与盖子放入沸水中煮 **2–3 分钟**，取出晾干。

### 🍯 2. 分开煮糖浆与果肉  
1. 用滤网把草莓果肉捞出，糖浆单独倒回锅里。  
2. 中大火煮至糖浆 **明显变浓稠**，温度约 **105°C**（页面原说明）

### 🍓 3. 加入草莓果肉  
1. 倒回草莓果肉，中小火煮至再次沸腾。  
2. 再煮 **5 分钟**。  
3. 做“凝胶测试”：  
   - 滴一点果酱在冷盘上  
   - 放凉后用手指划线  
   - 若线条分明 → 完成

### 🫙 4. 装瓶  
趁热装瓶、旋紧、倒扣放凉。  
冷却后放冰箱保存。

---

# 🌟 **lime 替代 lemon 的风味变化**
- 香气更清新、带一点青皮的清爽味  
- 酸度略高，能让草莓味更“跳”  
- 颜色依旧是宝石红，不受影响  

如果你喜欢草莓冷意面的甜品风格，这个 lime 版本会更适合，味道更轻盈。
