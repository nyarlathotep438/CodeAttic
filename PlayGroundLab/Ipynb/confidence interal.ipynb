```python
import math  
  
def sample_std_dev(numbers):  
    # 计算平均值  
    mean = sum(numbers) / len(numbers)  
      
    # 计算每个数值与平均值的差的平方和  
    variance = sum((x - mean) ** 2 for x in numbers) / (len(numbers) - 1)  
      
    # 计算标准差  
    std_dev = math.sqrt(variance)  
      
    return std_dev
```

```python
numbers = [98.84,125.24,111.64,108.65,91.28,105.68,100.49,83.56,90.45,102.33,109.86,88.50,93.50,105.32]
print(sample_std_dev(numbers))
```

```python
numbers = [9,8.5,14,10.2,11]
print(sample_std_dev(numbers))
```

```python
import numpy as np  
  
# 假设你有一个名为samples的numpy数组，其中包含你的样本数据  
samples = np.array([9,8.5,14,10.2,11])  # 替换为你的样本数据  
  
# 使用numpy的std函数来计算标准差  
sd = np.std(samples)  
  
print(f"The standard deviation is {sd}")
```

```python
import math  
  
# 已知数据  
n1, n2 = 15, 18  # 两个样本的大小  
x1_bar, x2_bar = 115.4, 117.2  # 两个样本的均值  
s1, s2 = 13, 17  # 两个样本的标准差  
z = 2.145  # 95%置信水平的z分数  
  
# 计算标准误差  
SE = math.sqrt((s1 ** 2 / n1) + (s2 ** 2 / n2))  
  
# 计算置信区间的上下限  
lower_bound = (x1_bar - x2_bar) - z * SE  
upper_bound = (x1_bar - x2_bar) + z * SE  
  
# 输出结果  
print(f"两年间学生身高均值差异的95%置信区间为 ({lower_bound:.2f}, {upper_bound:.2f})")  
  
# 判断是否存在身高均值差异  
if lower_bound < 0 < upper_bound:  
    print("由于置信区间包含0，因此不能断定两年间学生的身高均值存在显著差异。")  
else:  
    print("两年间学生的身高均值存在显著差异。")
```
