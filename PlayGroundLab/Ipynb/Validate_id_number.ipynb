```python
def validate_id_number(id_number: str) -> bool:
    """
    校验中国大陆18位身份证号码是否合法
    按 GB11643-1999 标准计算校验码
    """

    # 身份证必须是18位
    if len(id_number) != 18:
        return False

    # 前17位必须是数字
    if not id_number[:17].isdigit():
        return False

    # 加权因子（固定）
    weights = [7, 9, 10, 5, 8, 4, 2, 1, 6, 3, 7, 9, 10, 5, 8, 4, 2]

    # 校验码映射表（余数 → 校验码）
    check_map = {
        0: '1',
        1: '0',
        2: 'X',
        3: '9',
        4: '8',
        5: '7',
        6: '6',
        7: '5',
        8: '4',
        9: '3',
        10: '2'
    }

    # 计算加权和
    total = sum(int(num) * weight for num, weight in zip(id_number[:17], weights))

    # 取模得到应有的校验码
    expected_check_digit = check_map[total % 11]

    # 实际校验码（可能是小写 x）
    actual_check_digit = id_number[17].upper()

    return expected_check_digit == actual_check_digit
```

```python
# 示例测试
tests = [
    "11010519491231002X",  # 合法示例
    "110105194912310021",  # 校验码错误
    "123456789012345",     # 长度不对
]

for t in tests:
    print(t, "=>", validate_id_number(t))
```
