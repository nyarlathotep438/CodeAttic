```python
from scipy.stats import pearsonr  
  
data1 = [4,5,9,10,12,15,18,25,30]  
data2 = [9.2,8.4,7.6,8,6.5,7.4,7,6.2,5]  
    
r, _ = pearsonr(data1, data2)  
  
print(f"Pearson's correlation coefficient: {r}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
filepath = "D:/work1.csv"

data = pd.read_csv(filepath)
```

```python
data.head()
```

```python
sns.regplot(x=data['Temperature'], y=data['Ice Cream Sales'])
```

```python
from scipy.stats import pearsonr  
  
data1 = data['Temperature']
data2 = data['Ice Cream Sales']
  
r, _ = pearsonr(data1, data2)  
  
print(f"Pearson's correlation coefficient: {r}")
print(f"coefficient of determination: {r*r}")
```

```python
import numpy as np  
from scipy import stats  
   
x = data['Temperature']
y = data['Ice Cream Sales']  
  
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)  
 
print(f"线性回归方程为：y = {slope}x + {intercept}")  
print(f"其中，斜率为 {slope}，截距为 {intercept}。")
```
