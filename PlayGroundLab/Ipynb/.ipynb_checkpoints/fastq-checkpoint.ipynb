```python
import matplotlib.pyplot as plt  
  
# Prepare the data  
technologies = ['CRISPR', 'ZFNs', 'TALENs']  
editing_efficiency = [47, 32, 28]  # In percentages  
off_target_rate = [23.21, 38.42, 57.4]  # In percentages  
  
# Set the width and location for the bars  
bar_width = 0.35  
index = range(len(technologies))  
  
# Set up the plot  
plt.figure(figsize=(12, 6))  
#plt.title('Comparison of Gene Editing Efficiency and Off-target Rate')  
plt.xlabel('Gene Editing Tools')  
plt.ylabel('Percentage (%)')  
  
# Plot the data  
p1 = plt.bar(index, editing_efficiency, bar_width, color='green', label='Gene Editing Efficiency')  
p2 = plt.bar([i + bar_width for i in index], off_target_rate, bar_width, color='red', label='Off-target Rate')  
  
# Add labels, legends, and annotations  
plt.xticks([i + bar_width / 2 for i in index], technologies)  
plt.legend()  
plt.figtext(0.5, 0.01, 'fig3. Comparison of gene editing efficiency and off-target rate among different gene editing tools', ha='center', fontsize=11)  
  
# Save the figure  
plt.savefig('gene_editing_comparison.png')  
  
# Optionally display the figure (uncomment the line below if needed)  
# plt.show()
```

```python
import matplotlib.pyplot as plt  
  
# 准备数据  
technologies = ['CRISPR', 'ZFNs', 'TALENs']  
efficiencies = [47.32, 32.28, 28.94]  
  
# 设置图表  
plt.figure(figsize=(10, 6))  # 设置图表大小  
#plt.title('Gene Editing Efficiencies of Different Technologies')  # 设置图表标题  
plt.xlabel('Gene Editing Technology')  # 设置x轴标签  
plt.ylabel('Editing Efficiency (%)')  # 设置y轴标签  
  
# 柱状图的宽度  
bar_width = 0.4  # 设置柱宽度为0.4  
  
# 索引位置调整以适应柱宽度，确保x轴坐标位于柱的居中位置  
index = range(len(technologies))  # 原始索引  
  
# 绘制柱状图，选择一个浅蓝色  
plt.bar(index, efficiencies, width=bar_width, color='blue')  
  
  
plt.xticks(index, technologies) 
plt.figtext(0.5, 0.01, 'fig2. Comparison of gene editing efficiency among different gene editing tools', ha='center', fontsize=11)  
  
# 添加数据标签  
for idx, val in enumerate(efficiencies):  
    plt.text(idx, val + 1, f'{val:.2f}%', ha='center')  
  
# 保存并显示图表  
plt.savefig('gene_editing_efficiencies.png')  # 保存图表为PNG文件  
plt.show()  # 显示图表
```

```python
import matplotlib.pyplot as plt  
  
# 基因序列及其反义链  
sequence = "ATTTGGAAAATGGATATAATCAAAATCATGGAAGAAAGTTTGTACAGGGTA"  
antisense = "TAAACCTTTTACCTATATTAGTTTTAGTACCTTCTTTCAAACATGTCCCAT"  
  
# 碱基与颜色的映射，修改了黄色为暗黄色  
base_colors = {  
    'A': 'red',  
    'T': 'green',  
    'C': 'blue',  
    'G': '#8B4513'  # 暗黄色  
}  
  
# 绘图设置  
fig, ax = plt.subplots(figsize=(15, 3))  # 减小了图形的高度  
  
# 为了使两行贴近，我们减小了y轴上的间隔  
y_positions = [0.5, -0.5]  # 修改了y轴位置，使两行更贴近  
  
# 绘制碱基对  
for i, (base, antisense_base) in enumerate(zip(sequence, antisense)):  
    # 绘制正义链碱基  
    ax.text(i, y_positions[0], base, color=base_colors[base], fontsize=14, ha='center', va='center')  
    # 绘制反义链碱基  
    ax.text(i, y_positions[1], antisense_base, color=base_colors[antisense_base], fontsize=14, ha='center', va='center')  
  
# 在序列两端添加"..."以表示基因不全  
ax.text(-1, y_positions[0], '...', fontsize=14, va='center')  
ax.text(len(sequence), y_positions[0], '...', fontsize=14, va='center')  
ax.text(-1, y_positions[1], '...', fontsize=14, va='center')  
ax.text(len(antisense), y_positions[1], '...', fontsize=14, va='center')  
  
# 设置坐标轴范围和属性  
ax.set_xlim(-2, len(sequence) + 1)  
ax.set_ylim(-1.5, 1.5)  # 修改了y轴范围，以适应新的y轴位置  
ax.axis('off')  # 关闭坐标轴显示  

#plt.figtext(0.5, 0.01, 'fig1.CRISPR edits AMD-related genes', ha='center', fontsize=11)  
  
# 显示图形  
plt.tight_layout()  # 调整子图参数，使之填充整个图像区域  
plt.savefig('rs1061170.png') 
plt.show()
```
