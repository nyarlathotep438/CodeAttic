```python
import torch
import numpy as np
```

```python
np_data = np.arange(6).reshape((2,3))
torch_data = torch.from_numpy(np_data)
tensor2array = torch_data.numpy()
```

```python
print(
    '\nnumpy',np_data,
    '\ntorch',torch_data,
    '\ntensor2array',tensor2array,
)
```

```python
# abs
data = [-1, -2, 1, 2]
tensor = torch.FloatTensor(data) #32-bit floating point
```

```python
print(
    '\nabs',
    '\nnumpy',np.abs(data),
    '\ntorch',torch.abs(tensor),
)
```
