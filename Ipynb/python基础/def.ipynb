```python
def sale_car(price, color='red', brand='carmy', is_second_hand=True):
    print('price', price,
          'color', color,
          'brand', brand,
          'is_second_hand', is_second_hand,)
```

```python
sale_car(1000)
```

```python
def report(name, *grades):
    total_grade = 0
    for grade in grades:
        total_grade += grade
    print(name, 'total grade is ', total_grade)
```

```python
report('Mike', 8, 9)
```

```python
report('Mike', 8, 9, 10)
```

```python
def portrait(name, **kw):
    print('name is', name)
    for k,v in kw.items():
        print(k, v)
```

```python
portrait('Mike', age=24, country='China', education='bachelor')
```

```python
APPLE = 100 # 全局变量
a = None
def fun():
    global a    # 使用之前在全局里定义的 a
    a = 20      # 现在的 a 是全局变量了
    return a+100

print(APPLE)    # 100
print('a past:', a)  # None
fun()
print('a now:', a)   # 20
```
