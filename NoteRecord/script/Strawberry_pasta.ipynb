# 🔧 **制作步骤（关键在温度与打发）**

## **1. 融化白巧克力（关键步骤）**
- 将白巧克力隔水加热或微波炉 10–15 秒一轮  
- 搅拌至完全顺滑  
- **放凉至不烫手（约 35–40°C）**

> 温度太高会导致后面和奶油混合时“油水分离”。

---

## **2. 打发淡奶油**
- 淡奶油 + 一小撮盐  
- 打到 **6 分发**：  
  - 有纹路  
  - 但仍然能流动  
  - 提起打蛋器会形成柔软的弯钩

> 不能打太硬，否则混合后会颗粒感明显。

---

## **3. 混合（最关键的步骤）**
- 将 **1/3 打发奶油** 加入白巧克力中，快速搅匀  
  → 这是“调温”，让两者质地接近  
- 再把混合物倒回剩下的奶油里  
- 用刮刀 **轻轻翻拌**（fold），不要画圈搅拌

> 这样做出来的慕斯会非常细腻、轻盈、稳定。

---

## **4. 调整稠度（可选）**
如果你想让慕斯更适合挤花：

- 加 10–20 ml 冷牛奶  
- 或者多打 10 秒淡奶油

如果你想让慕斯更“像云朵”：

- 淡奶油打到 5 分发即可（更轻盈）

# 🍓 为什么超市果酱是固体？
超市果酱通常：
- 含 **高果胶**（天然或添加）
- 含 **高糖**（为了保存）
- 经过 **长时间熬煮**（水分蒸发 → 变稠）
- 目标是“可涂抹”，不是“可流动”

所以它自然是固体状、浓稠、甚至可以立起来。

---

# 🍓 餐厅草莓酱为什么是液体？
餐厅的草莓酱通常是：
- **短时间加热**（保持果香）
- **低糖**（不需要保存）
- **不加果胶**
- **加入柠檬汁或水调整流动性**

它更像“草莓汁 + 果肉”的混合物。

---

# 🍓 在家做“餐厅级草莓酱汁”的方法（最推荐）
这个版本适合淋在甜品、意面、冰淇淋、煎饼上。

## 🧂 配料（约 200 ml）
- 草莓 150 g（新鲜或冷冻）
- 糖 15–25 g（按草莓甜度调整）
- 柠檬汁 1–2 茶匙
- 水 1–2 汤匙（决定流动性）
- 可选：香草精 2–3 滴

---

# 🔥 做法（5 分钟）
1. **草莓切小块**，放入小锅。
2. 加入 **糖 + 柠檬汁 + 水**。
3. 中小火加热 **2–4 分钟**，直到草莓开始出汁、变软。
4. 关火，用勺子轻压几下（保持果肉感）。
5. 如果你想更顺滑：  
   → 用料理棒打几秒  
   → 或过滤掉籽，得到丝滑酱汁

冷却后会稍微变稠，但仍然是流体。

# 🍫🍓 **白巧克力草莓意面卷（White Chocolate Strawberry Pasta Roll）**

## 🎯 设计理念（为什么它可行）
- **意面 = 弹性结构体**  
  冷却后略带韧性，非常适合卷起来，不会像可丽饼那样软塌。
- **白巧克力奶油 = 黏合剂**  
  冷却后会轻微凝固，能把卷固定住。
- **草莓酱 = 酸度与亮点**  
  切开后形成“红白相间”的视觉效果。
- **整体风味 = 草莓白巧克力千层 × 意面寿司**

---

# 🧪 配方（2–3 卷）

### **A. 意面底（结构层）**
- 意面（建议用 linguine 或 tagliatelle）…… 80–100 g（干重）
- 盐……少许  
- 冷水 + 冰块（冷却用）

### **B. 白巧克力奶油（黏合层）**
- 白巧克力……50 g  
- 淡奶油……80 ml  
- 马斯卡彭……40 g（或希腊酸奶替代）  
- 柠檬皮屑……少许  
- 香草精……2–3 滴  

### **C. 草莓酱（色彩与酸度）**
（你之前已经做过，我给你一个卷用版本）
- 草莓……80 g  
- 糖……8–12 g  
- 柠檬汁……1 茶匙  
- 水……1 汤匙  
- 盐……一小撮  

### **D. 装饰（可选）**
- 冻干草莓碎  
- 白巧克力刨花  
- 新鲜草莓片

# 🔥 制作步骤（核心流程）

## **1. 煮意面并冷却**
- 煮到 **比 al dente 再硬一点点**（因为要卷）  
- 捞出后立即放入冰水中冷却  
- 沥干，拌一点点淡味油防粘  

👉 冷却后的意面会更有弹性，适合卷。

---

## **2. 做白巧克力奶油**
- 小火加热淡奶油至微冒泡  
- 离火加入白巧克力搅匀  
- 稍微冷却后加入马斯卡彭、柠檬皮屑、香草精  
- 冷藏 10–15 分钟让它稍微变稠  

👉 质地要像“可抹的奶油”，不能太稀。

---

## **3. 做草莓酱（卷用版本）**
- 草莓 + 糖 + 柠檬汁 + 水，小火煮 2–3 分钟  
- 压碎部分草莓  
- 加一小撮盐  
- 冷却备用  

👉 酱要偏稠，这样切开时不会流得太厉害。

---

## **4. 组装（关键步骤）**
在案板上铺一张保鲜膜：

1. 把意面**横向铺成一层薄薄的“面片”**（像寿司饭那样）  
2. 在上面抹一层薄薄的白巧克力奶油  
3. 在靠近你的一侧放一条草莓酱（像寿司卷的馅）  
4. 用保鲜膜辅助，把意面卷起来  
5. 卷紧后放冰箱冷藏 20–30 分钟定型  

👉 冷藏定型是成功的关键。

---

## **5. 切片与摆盘**
- 用锋利的刀切成 2–3 cm 厚的卷  
- 顶部点一点草莓酱  
- 撒冻干草莓碎 + 白巧克力刨花  

切开后你会看到：

- 外层：淡黄色意面  
- 中层：白巧克力奶油  
- 内芯：鲜红草莓酱  

视觉效果非常像甜点寿司。

---

# 🍽️ 风味表现（你会喜欢的点）
- **白巧克力的甜 + 草莓的酸** → 经典组合  
- **意面的弹性** → 让甜点有“咀嚼感”，不腻  
- **冷藏后的奶油微凝固** → 口感像慕斯  
- **切面非常漂亮** → 适合拍照、适合做成 notebook 的封面图
