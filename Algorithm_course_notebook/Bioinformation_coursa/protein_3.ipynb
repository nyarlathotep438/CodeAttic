```python
# -----------------------------
# Spectrum Graph Construction
# -----------------------------

# 1. 标准 20 种氨基酸的整数质量表
mass_table = {
    57: "G",
    71: "A",
    87: "S",
    97: "P",
    99: "V",
    101: "T",
    103: "C",
    113: "I/L",
    114: "N",
    115: "D",
    128: "K/Q",
    129: "E",
    131: "M",
    137: "H",
    147: "F",
    156: "R",
    163: "Y",
    186: "W"
}

# 2. 读取输入（你可以改成手动输入或直接写死）
# 例如：57 71 154 185 301 332 415 429 486
spectrum = list(map(int, input("Enter spectrum: ").split()))

# 3. 把 0 加入 spectrum（代表空前缀）
spectrum.append(0)

# 4. 排序
spectrum = sorted(spectrum)

# 5. 构建图
edges = []

for u in spectrum:
    for v in spectrum:
        diff = v - u
        if diff in mass_table:
            aa = mass_table[diff]
            # 如果是 I/L 或 K/Q，你可以选择输出其中一个字母
            if "/" in aa:
                aa = aa.split("/")[0]
            edges.append(f"{u}->{v}:{aa}")

# 6. 输出结果
for e in edges:
    print(e)
```

```python
# Standard integer mass table for 20 amino acids
mass_table = {
    57: "G",
    71: "A",
    87: "S",
    97: "P",
    99: "V",
    101: "T",
    103: "C",
    113: "L",
    114: "N",
    115: "D",
    128: "Q",
    129: "E",
    131: "M",
    137: "H",
    147: "F",
    156: "R",
    163: "Y",
    186: "W"
}

def construct_graph(spectrum, mass_table):
    spectrum = sorted(spectrum)
    spectrum = [0] + spectrum  # add source mass 0

    graph = {m: [] for m in spectrum}

    for u in spectrum:
        for v in spectrum:
            diff = v - u
            if diff in mass_table:
                aa = mass_table[diff]
                if "/" in aa:
                    aa = aa.split("/")[0]  # choose one if ambiguous
                graph[u].append((v, aa))
    return graph

def find_path(graph, start, end):
    stack = [(start, [])]  # (current_node, path_edges)

    while stack:
        node, path = stack.pop()

        if node == end:
            return path

        for nxt, aa in graph[node]:
            stack.append((nxt, path + [aa]))

    return None  # no path found

def decode_ideal_spectrum(spectrum):
    graph = construct_graph(spectrum, mass_table)
    start = 0
    end = max(spectrum)

    path = find_path(graph, start, end)
    if path is None:
        return "No peptide found"

    return "".join(path)

spectrum = list(map(int, input("Enter spectrum: ").split()))
peptide = decode_ideal_spectrum(spectrum)
print(peptide)
```
