```python
import sys
sys.setrecursionlimit(10000)

def longest_common_subsequence(s1, s2):
    m, n = len(s1), len(s2)
    # 初始化DP表
    dp = [[0] * (n + 1) for _ in range(m + 1)]  
    
    # 填充DP表
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    
    # 回溯构造LCS
    lcs = []
    i, j = m, n
    while i > 0 and j > 0:
        if s1[i-1] == s2[j-1]:
            lcs.append(s1[i-1])
            i -= 1
            j -= 1
        elif dp[i-1][j] > dp[i][j-1]:
            i -= 1
        else:
            j -= 1
    
    return ''.join(reversed(lcs))

# 测试验证
print(longest_common_subsequence("AGACTG", "GTACGA"))
```

```python
import sys
from collections import defaultdict

def longest_path_dag(start, end, edges):
    # Build adjacency list with weights
    graph = defaultdict(list)
    nodes = set()
    for u, v, w in edges:
        graph[u].append((v, w))
        nodes.add(u)
        nodes.add(v)

    # Initialize distances and predecessors
    dist = {node: -float("inf") for node in nodes}
    dist[start] = 0
    prev = {node: None for node in nodes}

    # Since nodes are already in topological order (increasing labels)
    for u in sorted(nodes):
        if dist[u] != -float("inf"):
            for v, w in graph[u]:
                if dist[v] < dist[u] + w:
                    dist[v] = dist[u] + w
                    prev[v] = u

    # Reconstruct path
    path = []
    cur = end
    while cur is not None:
        path.append(cur)
        cur = prev[cur]
    path.reverse()

    return dist[end], path


if __name__ == "__main__":
    # Sample Input
    input_data = """0 4
0 1 7
0 2 4
2 3 2
1 4 1
3 4 3"""
    
    lines = input_data.strip().split("\n")
    start, end = map(int, lines[0].split())
    edges = []
    for line in lines[1:]:
        u, v, w = map(int, line.split())
        edges.append((u, v, w))

    length, path = longest_path_dag(start, end, edges)
    print(length)
    print(" ".join(map(str, path)))
```

```python
def find_all_LCSs(x,y):

    m = len(x)
    n = len(y)

    # Build the DP table
    dp_table = [[0] * (n+1) for _ in range(m+1)]

    for i in range(1, m+1):
        for j in range(1, n+1):
            
            if x[i-1] == y[j-1]:
                dp_table[i][j] = dp_table[i-1][j-1] + 1
            else:
                dp_table[i][j] = max(dp_table[i-1][j], dp_table[i][j-1])

    # Helper function to reconstruct LCS and backtrack through all solutions
    def backtrack_all(i,j):
        
        if i == 0 or j == 0:
            return {""}
        
        if x[i - 1] == y[j - 1]:
            subsequences = backtrack_all(i-1, j-1)
            return {s + x[i - 1] for s in subsequences}
        
        else:
            result_set = set()

        # Explore the path from the top(excluding X[i-1])

        if dp_table[i - 1][j] >= dp_table[i][j - 1]:
            result_set.update(backtrack_all(i-1,j))
        
        # Explore the path from the left(excluding Y[i-1])

        if dp_table[i][j - 1] >= dp_table[i - 1][j]:
            result_set.update(backtrack_all(i,j-1))

        return result_set
    
    return backtrack_all(m,n)
```

```python
import math
def Manhattan_Tourist_Problem_path(m,n):
    #math.comb(m+n, n)
    path_amount = math.factorial(m + n) / (math.factorial(m) * math.factorial(n))
    path_amount = int(path_amount)
    return path_amount
```

```python
def dynamic_programming_change(money: int, coins: list[int]):
    min_num_coins = [0] * (money + 1)
    for coin in coins:
        if coin < money:
            min_num_coins[coin] = 1

    for m in range(1, money + 1):
        min_num_coins[m] = min([min_num_coins[m - coin] if m - coin >= 0 else math.inf for coin in coins]) + 1

    return min_num_coins[money]
```

```python
def manhattan_tourist(n, m, Down, Right):
    # Initialize DP matrix
    s = [[0] * (m + 1) for _ in range(n + 1)]

    # Fill in first column using Down matrix
    for i in range(1, n + 1):
        s[i][0] = s[i - 1][0] + Down[i - 1][0]

    # Fill in first row using Right matrix
    for j in range(1, m + 1):
        s[0][j] = s[0][j - 1] + Right[0][j - 1]

    # Fill in rest of DP matrix
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            s[i][j] = max(
                s[i - 1][j] + Down[i - 1][j],  # coming from above
                s[i][j - 1] + Right[i][j - 1]  # coming from left
            )
    return s[n][m]
```
