```python
def find_pattern_positions(pattern, genome):
    """
    在Genome字符串中查找Pattern的所有起始位置
    参数:
        pattern (str): 要搜索的模式字符串
        genome (str): 被搜索的基因组字符串
    返回:
        str: 以空格分隔的起始位置字符串
    """
    positions = []
    start = 0
    pattern_length = len(pattern)
    
    while start < len(genome):
        # 在genome中查找pattern的下一个出现位置
        pos = genome.find(pattern, start)
        if pos == -1:
            break  # 没有更多匹配项
        positions.append(str(pos))  # 记录找到的位置
        start = pos + 1  # 从下一个位置继续搜索（处理重叠匹配）
    
    return " ".join(positions)
```

```python
def FindClumps(Text, k, L, t):
    # 初始化结果集
    patterns = []
    n = len(Text)
    
    # 遍历所有可能的滑动窗口
    for i in range(n - L + 1):
        window = Text[i:i+L]  # 当前窗口序列
        
        # 构建k-mer频率表
        freq_map = {}
        for j in range(len(window) - k + 1):
            kmer = window[j:j+k]
            freq_map[kmer] = freq_map.get(kmer, 0) + 1
        
        # 筛选满足频次条件的k-mer
        for kmer, count in freq_map.items():
            if count >= t and kmer not in patterns:
                patterns.append(kmer)
    
    return patterns
```

```python
def count_substring(sequence, substring):
    count = 0
    sub_len = len(substring)
    # 遍历所有可能起始位置（考虑重叠）
    for i in range(len(sequence) - sub_len + 1):
        if sequence[i:i+sub_len] == substring:
            count += 1
    return count

# 输入数据
dna_sequence = "ATGACTTCGCTGTTACGCGC"
target = "CGC"

# 计算结果
result = count_substring(dna_sequence, target)
print(f"子串 '{target}' 出现次数: {result}")
```
