```python
def translate_rna_to_polypeptide(rna_sequence, codon_dict):
    """将RNA序列翻译为多肽链，遇到终止密码子立即停止"""
    polypeptide = []  # 存储氨基酸序列
    
    # 每次取3个碱基作为密码子
    for i in range(0, len(rna_sequence) - 2, 3):  # 确保有足够的碱基组成密码子
        codon = rna_sequence[i:i+3]
        
        # 查找密码子对应的氨基酸
        amino_acid = codon_dict.get(codon)
        
        if amino_acid is None:  # 遇到无效密码子
            print(f"警告：遇到无效密码子 '{codon}'，翻译终止")
            break
        elif amino_acid == 'STOP':  # 遇到终止密码子
            print(f"在位置 {i}-{i+2} 遇到终止密码子 '{codon}'，翻译终止")
            break
        else:
            polypeptide.append(amino_acid)
    
    return ''.join(polypeptide)  # 将氨基酸列表连接成字符串
```

```python
def LinearSpectrum(Peptide, AminoAcidMass):
    """
    计算线性肽的质谱（优化版）
    
    参数:
    Peptide (str): 氨基酸序列字符串
    AminoAcidMass (dict): 氨基酸到质量的映射字典
    
    返回:
    list: 排序后的质谱列表
    """
    n = len(Peptide)
    PrefixMass = [0] * (n + 1)
    
    # 计算前缀质量（使用字典直接验证）
    for i in range(1, n + 1):
        aa = Peptide[i - 1]
        if aa not in AminoAcidMass:
            raise ValueError(f"氨基酸 '{aa}' 的质量未定义")
        PrefixMass[i] = PrefixMass[i - 1] + AminoAcidMass[aa]
    
    # 生成质谱
    LinearSpectrum = [0]
    for i in range(n):
        for j in range(i + 1, n + 1):
            mass = PrefixMass[j] - PrefixMass[i]
            LinearSpectrum.append(mass)
    
    return sorted(LinearSpectrum)
```

```python
def CyclicSpectrum(Peptide, AminoAcidMass):
    """
    计算环肽的质谱（修复版）
    
    参数:
    Peptide (str): 氨基酸序列字符串
    AminoAcidMass (dict): 氨基酸到质量的映射字典
    
    返回:
    list: 排序后的唯一质谱值列表
    """
    n = len(Peptide)
    total_mass = sum(AminoAcidMass[aa] for aa in Peptide)
    
    # 初始化前缀质量数组
    PrefixMass = [0] * (n + 1)
    for i in range(1, n + 1):
        aa = Peptide[i - 1]
        if aa not in AminoAcidMass:
            raise ValueError(f"氨基酸 '{aa}' 的质量未定义")
        PrefixMass[i] = PrefixMass[i - 1] + AminoAcidMass[aa]
    
    # 初始化质谱列表（只包含0）
    spectrum_list = [0]
    
    # 1. 添加所有线性片段（包括整个肽）
    for i in range(n):
        for j in range(i + 1, n + 1):
            mass = PrefixMass[j] - PrefixMass[i]
            spectrum_list.append(mass)
    
    # 2. 正确添加跨越末端的片段（仅非连续部分）
    for i in range(1, n):  # 从1开始，避免重复0
        for j in range(i + 1, n):
            # 计算跨越末端的片段：从j到末尾 + 从开头到i-1
            mass = (PrefixMass[n] - PrefixMass[j]) + PrefixMass[i]
            spectrum_list.append(mass)
    
    # 去重并排序
    return sorted(spectrum_list)
```

```python
from collections import Counter

 

AMINO_ACID_MASSES = [

    57, 71, 87, 97, 99, 101, 103, 113, 114, 115,

    128, 129, 131, 137, 147, 156, 163, 186

]

 

def cyclic_spectrum(peptide):

    prefix_mass = [0]

    for m in peptide:

        prefix_mass.append(prefix_mass[-1] + m)

    peptide_mass = prefix_mass[-1]

    spectrum = [0]

    for i in range(len(peptide)):

        for j in range(i+1, len(peptide)+1):

            spectrum.append(prefix_mass[j] - prefix_mass[i])

            if i > 0 and j < len(peptide):

                spectrum.append(peptide_mass - (prefix_mass[j] - prefix_mass[i]))

    return sorted(spectrum)

 

def linear_spectrum(peptide):

    prefix_mass = [0]

    for m in peptide:

        prefix_mass.append(prefix_mass[-1] + m)

    spectrum = [0]

    for i in range(len(peptide)):

        for j in range(i+1, len(peptide)+1):

            spectrum.append(prefix_mass[j] - prefix_mass[i])

    return sorted(spectrum)

 

def is_consistent(peptide, spectrum):

    peptide_spectrum = linear_spectrum(peptide)

    peptide_counter = Counter(peptide_spectrum)

    spectrum_counter = Counter(spectrum)

    for mass in peptide_counter:

        if peptide_counter[mass] > spectrum_counter[mass]:

            return False

    return True

 

def expand(peptides):

    return [peptide + [mass] for peptide in peptides for mass in AMINO_ACID_MASSES]

 

def cyclopeptide_sequencing(spectrum):

    peptides = [[]]

    final_peptides = []

    parent_mass = max(spectrum)

 

    while peptides:

        peptides = expand(peptides)

        new_peptides = []

        for peptide in peptides:

            mass = sum(peptide)

            if mass == parent_mass:

                if cyclic_spectrum(peptide) == sorted(spectrum):

                    if peptide not in final_peptides:

                        final_peptides.append(peptide)

            elif is_consistent(peptide, spectrum):

                new_peptides.append(peptide)

        peptides = new_peptides

 

    return final_peptides

 

 

def read_spectrum_from_file(filename):

    with open(filename, 'r') as file:

        line = file.readline().strip()

        return list(map(int, line.split()))

 

# Read the spectrum from your .txt file

spectrum = read_spectrum_from_file(r"C:\Users\Admin\Downloads\dataset_30217_6.txt")

#Spectrum = "0 99 113 114 128 227 257 299 355 356 370 371 484"

# Now run the algorithm

result = cyclopeptide_sequencing(spectrum)

 

# Print results

for peptide in result:

    print("-".join(map(str, peptide)))
```
