```python
def StringComposition(k, Text):
    kmers = [Text[i:i+k] for i in range(len(Text) - k + 1)]
    return ' '.join(kmers)
```

```python
def genome_path_string(kmers):
    """Given an ordered list of overlapping k-mers, return the spelled string."""
    if not kmers:
        return ""
    return kmers[0] + "".join(k[-1] for k in kmers[1:])

# Your input
kmers_str = """ATGCG GCATG CATGC AGGCA GGCAT GGCAC"""
# Split into list
kmers = kmers_str.split()

# Reconstruct string
result = genome_path_string(kmers)
print(result)
```

```python
def OverlapGraph(kmers):
    """Construct the overlap graph of kmers as an adjacency list."""
    overlap_graph = {}
    k = len(kmers[0]) if kmers else 0  # 假设所有k-mers长度相同
    
    for current_kmer in kmers:
        suffix = current_kmer[1:]  # 后k-1个字符
        successors = []
        
        for other_kmer in kmers:
            if current_kmer == other_kmer:
                continue  # 跳过自环
            if other_kmer.startswith(suffix):  # 检查other_kmer的前k-1是否匹配current_kmer的后k-1
                successors.append(other_kmer)
        
        if successors:
            overlap_graph[current_kmer] = " ".join(successors)
    
    return overlap_graph
```

```python
from collections import defaultdict, deque

kmers_str = "AAAT AATG ACCC ACGC ATAC ATCA ATGC CAAA CACC CATA CATC CCAG CCCA CGCT CTCA GCAT GCTC TACG TCAC TCAT TGCA"
kmers = kmers_str.split()

def build_overlap_graph(kmers):
    graph = defaultdict(set)
    in_degree = defaultdict(int)
    out_degree = defaultdict(int)
    
    for kmer in kmers:
        prefix = kmer[:-1]  # 前 k-1 个字符（例如 "AAA" 的 prefix 是 "AA"）
        suffix = kmer[1:]   # 后 k-1 个字符（例如 "AAA" 的 suffix 是 "AA"）
        graph[prefix].add(suffix)
        out_degree[prefix] += 1
        in_degree[suffix] += 1
    
    return graph, in_degree, out_degree

def is_eulerian_path_possible(graph, in_degree, out_degree):
    # 检查是否弱连通（忽略方向）
    if not is_weakly_connected(graph):
        return False
    
    start_nodes = 0
    end_nodes = 0
    
    for node in set(in_degree.keys()).union(set(out_degree.keys())):
        if out_degree[node] - in_degree.get(node, 0) == 1:
            start_nodes += 1
        elif in_degree.get(node, 0) - out_degree[node] == 1:
            end_nodes += 1
        elif out_degree[node] != in_degree.get(node, 0):
            return False  # 某个节点的出度和入度差超过1，不满足条件
    
    # 欧拉路径条件：start_nodes=1 且 end_nodes=1
    # 欧拉循环条件：start_nodes=0 且 end_nodes=0（所有节点出度=入度）
    return (start_nodes == 1 and end_nodes == 1) or (start_nodes == 0 and end_nodes == 0)

def is_weakly_connected(graph):
    if not graph:
        return True
    
    # 忽略方向，检查连通性
    adjacency = defaultdict(set)
    all_nodes = set()
    
    for u in graph:
        all_nodes.add(u)
        for v in graph[u]:
            all_nodes.add(v)
            adjacency[u].add(v)
            adjacency[v].add(u)  # 忽略方向
    
    if not all_nodes:
        return True
    
    # BFS 检查连通性
    visited = set()
    start_node = next(iter(all_nodes))
    queue = deque([start_node])
    visited.add(start_node)
    
    while queue:
        u = queue.popleft()
        for v in adjacency[u]:
            if v not in visited:
                visited.add(v)
                queue.append(v)
    
    return len(visited) == len(all_nodes)

# 构建重叠图
graph, in_degree, out_degree = build_overlap_graph(kmers)

# 检查欧拉路径是否存在
if is_eulerian_path_possible(graph, in_degree, out_degree):
    print("欧拉路径存在！")
else:
    print("欧拉路径不存在！")

# 打印图结构（可选）
print("\n重叠图结构：")
for node in sorted(graph.keys()):
    print(f"{node} -> {', '.join(sorted(graph[node]))}")
```

```python
from collections import defaultdict

def parse_adjacency_list(input_str):
    graph = {}
    for line in input_str.strip().split('\n'):
        if not line.strip():
            continue
        u, neighbors = line.split(': ')
        u = int(u)
        neighbors = list(map(int, neighbors.split())) if neighbors else []
        graph[u] = neighbors
    return graph

def find_eulerian_circuit(graph):
    stack = []
    path = []
    adj = defaultdict(list)
    for u in graph:
        adj[u] = list(graph[u])
    start = next(iter(adj))
    stack.append(start)
    while stack:
        current = stack[-1]
        if adj[current]:
            next_node = adj[current].pop()
            stack.append(next_node)
        else:
            path.append(stack.pop())
    return path[::-1]

# 输入邻接表字符串
input_str = """0: 3
1: 0
2: 1 6
3: 2
4: 2
5: 4
6: 5 8
7: 9
8: 7
9: 6
"""

# 解析并计算欧拉回路
graph = parse_adjacency_list(input_str)
circuit = find_eulerian_circuit(graph)
print(" ".join(map(str, circuit)))
```

```python
from collections import defaultdict, deque

def eulerian_path_from_string(graph_str):
    # Parse the input string into a graph (adjacency list)
    graph = defaultdict(list)
    for line in graph_str.split('\n'):
        if not line.strip():
            continue  # Skip empty lines
        node, edges = line.split(': ')
        node = int(node)  # Convert node to int (or keep as str if needed)
        edges = list(map(int, edges.split())) if edges else []
        graph[node] = edges
    
    # Step 1: Compute in-degree and out-degree for each node
    in_degree = defaultdict(int)
    out_degree = defaultdict(int)
    
    for u in graph:
        out_degree[u] = len(graph[u])
        for v in graph[u]:
            in_degree[v] += 1
    
    # Find start node (node with out_degree - in_degree == 1)
    start = None
    for u in graph:
        if out_degree[u] - in_degree[u] == 1:
            start = u
            break
    
    # If no unbalanced node, start can be any node (Eulerian cycle case)
    if start is None:
        # Pick any node with out_degree > 0 (or the first node if graph is not empty)
        start = next(iter(graph)) if graph else None
    
    # Step 2: Run Hierholzer's algorithm
    if start is None:
        return ""  # empty graph
    
    stack = [start]
    path = []
    adj = defaultdict(deque)
    for u in graph:
        adj[u] = deque(graph[u])  # Use deque for efficient popleft()
    
    while stack:
        current = stack[-1]
        if adj[current]:
            next_node = adj[current].popleft()
            stack.append(next_node)
        else:
            path.append(stack.pop())
    
    # Reverse to get the correct order and convert to string
    path_str = ' '.join(map(str, path[::-1]))
    return path_str

# Example usage
if __name__ == "__main__":
    input_str = """0: 1
1: 6 7
2: 0
4: 5
5: 1
6: 4
7: 9
8: 1
9: 8"""
    output_str = eulerian_path_from_string(input_str)
    print(output_str)
```

```python
def generate_binary_kmers(k):
    """生成所有长度为k的二进制字符串"""
    return [format(i, f'0{k}b') for i in range(2 ** k)]

def debruijn_graph_from_kmers(kmers):
    """从k-mers构建De Bruijn图"""
    graph = {}
    for kmer in kmers:
        prefix = kmer[:-1]
        suffix = kmer[1:]
        if prefix not in graph:
            graph[prefix] = []
        graph[prefix].append(suffix)
    return graph

def find_eulerian_cycle(graph):
    """使用Hierholzer算法寻找欧拉回路"""
    from collections import deque
    stack = deque()
    circuit = []
    # 选择一个起始节点（第一个节点）
    current_node = next(iter(graph.keys()))
    stack.append(current_node)
    
    while stack:
        current_node = stack[-1]
        # 如果当前节点还有未访问的出边
        if graph.get(current_node):
            # 取第一条出边
            next_node = graph[current_node].pop(0)
            stack.append(next_node)
        else:
            # 当前节点没有出边时，回溯
            circuit.append(stack.pop())
    
    # 将路径反转得到正确的欧拉回路
    circuit.reverse()
    return circuit

def assemble_string_from_cycle(cycle):
    """将欧拉回路中的节点拼接成字符串"""
    if not cycle:
        return ""
    result = cycle[0]
    for node in cycle[1:]:
        result += node[-1]
    return result

def universal_circular_string(k):
    """生成k-universal circular string"""
    if k <= 0:
        return ""
    if k == 1:
        return "01"  # 特殊处理k=1的情况
    
    # 1. 生成所有k位二进制字符串
    kmers = generate_binary_kmers(k)
    
    # 2. 构建De Bruijn图
    graph = debruijn_graph_from_kmers(kmers)
    
    # 3. 寻找欧拉回路
    cycle = find_eulerian_cycle(graph)
    
    # 4. 拼接序列
    s = assemble_string_from_cycle(cycle)
    
    # 5. 由于是循环字符串，取前2k个字符
    return s[:2**k]

# 测试
print(universal_circular_string(9))
```

```python
from collections import defaultdict, deque

def string_reconstruction(k, patterns):
    # Step 1: Build the De Bruijn graph
    graph = defaultdict(list)
    in_degree = defaultdict(int)
    out_degree = defaultdict(int)
    
    for pattern in patterns:
        if len(pattern) != k:
            continue  # Ensure all patterns are of length k
        prefix = pattern[:-1]
        suffix = pattern[1:]
        graph[prefix].append(suffix)
        out_degree[prefix] += 1
        in_degree[suffix] += 1
    
    # Step 2: Find the start node for Eulerian path
    start = None
    for node in graph:
        if out_degree[node] - in_degree[node] == 1:
            start = node
            break
    
    if start is None:
        # If no unbalanced node, start can be any node (Eulerian cycle case)
        start = next(iter(graph)) if graph else ""
    
    # Step 3: Hierholzer's algorithm to find Eulerian path
    stack = [start]
    path = []
    adj = defaultdict(deque)
    for u in graph:
        adj[u] = deque(graph[u])
    
    while stack:
        current = stack[-1]
        if adj[current]:
            next_node = adj[current].popleft()
            stack.append(next_node)
        else:
            path.append(stack.pop())
    
    eulerian_path = path[::-1]
    
    # Step 4: Reconstruct the string from the Eulerian path
    if not eulerian_path:
        return ""
    
    text = eulerian_path[0]
    for node in eulerian_path[1:]:
        text += node[-1]
    
    # Verify that the reconstructed string contains all k-mers
    # (This step is optional but can help catch errors)
    reconstructed_kmers = {text[i:i+k] for i in range(len(text) - k + 1)}
    if set(patterns) != reconstructed_kmers:
        return ""  # Reconstruction failed (graph had no valid Eulerian path)
    
    return text
```

```python
from collections import defaultdict

def calculate_min_edges(graph):
    in_degree = defaultdict(int)
    out_degree = defaultdict(int)
    
    # 统计出入度
    for node, neighbors in graph.items():
        out_degree[node] = len(neighbors)
        for neighbor in neighbors:
            in_degree[neighbor] += 1
    
    # 计算度差值
    delta = {}
    for node in set(out_degree.keys()).union(in_degree.keys()):
        delta[node] = out_degree.get(node, 0) - in_degree.get(node, 0)
    
    # 统计需要平衡的边数
    surplus = 0  # 总正差值（需要增加入边的总数量）
    deficit = 0  # 总负差值（需要增加出边的总绝对值）
    
    for node, d in delta.items():
        if d > 0:
            surplus += d
        elif d < 0:
            deficit += abs(d)
    
    # 最少需要添加的边数是两者的最大值
    min_edges = max(surplus, deficit)
    
    return min_edges, surplus, deficit

# 输入图（修正节点5的出边）
graph = {
    1: [2, 3, 5],
    2: [1, 4],
    3: [2, 5],
    4: [1, 2, 5],
    5: [3] 
}

min_edges, surplus, deficit = calculate_min_edges(graph)
print(f"最少需要添加 {min_edges} 条边")
```

```python
def parse_gapped_patterns(input_str):
    patterns = input_str.split()
    gapped_list = []
    for p in patterns:
        if '|' not in p:
            continue
        parts = p.split('|')
        if len(parts) != 2:
            continue
        first = parts[0].strip()
        second = parts[1].strip()
        gapped_list.append((first, second))
    return gapped_list

def build_deBruijn_graph(patterns):
    graph = {}
    for pattern in patterns:
        prefix = pattern[:-1]
        suffix = pattern[1:]
        if prefix not in graph:
            graph[prefix] = []
        graph[prefix].append(suffix)
    return graph

def find_eulerian_path(graph):
    out_degree = {}
    in_degree = {}
    
    for node in graph:
        out_degree[node] = len(graph[node])
        for neighbor in graph[node]:
            in_degree[neighbor] = in_degree.get(neighbor, 0) + 1
    
    all_nodes = set(out_degree.keys()) | set(in_degree.keys())
    
    start_node = None
    end_node = None
    for node in all_nodes:
        outdeg = out_degree.get(node, 0)
        indeg = in_degree.get(node, 0)
        if outdeg - indeg == 1:
            if start_node is not None:
                return None
            start_node = node
        if indeg - outdeg == 1:
            if end_node is not None:
                return None
            end_node = node

    if start_node is None:
        for node in graph:
            if out_degree.get(node, 0) > 0:
                start_node = node
                break
        if start_node is None:
            return None

    temp_graph = {}
    for node, edges in graph.items():
        temp_graph[node] = list(edges)
    
    stack = [start_node]
    path = []
    while stack:
        current = stack[-1]
        if current in temp_graph and temp_graph[current]:
            next_node = temp_graph[current].pop(0)
            stack.append(next_node)
        else:
            path.append(stack.pop())
    
    path = path[::-1]
    total_edges = sum(out_degree.values())
    if len(path) != total_edges + 1:
        return None
        
    return path

def path_to_string(path):
    if not path:
        return ""
    s = path[0]
    for i in range(1, len(path)):
        s += path[i][-1]
    return s

def StringSpelledByPatterns(patterns, k):
    if not patterns:
        return ""
    if len(patterns) == 1:
        return patterns[0]
    
    graph = build_deBruijn_graph(patterns)
    path = find_eulerian_path(graph)
    if path is None:
        return None
    return path_to_string(path)

def StringSpelledByGappedPatterns(gapped_patterns, k, d):
    n = len(gapped_patterns)
    if n < d + 1:
        return "there is no string spelled by the gapped patterns"

    FirstPatterns = [pair[0] for pair in gapped_patterns]
    SecondPatterns = [pair[1] for pair in gapped_patterns]

    PrefixString = StringSpelledByPatterns(FirstPatterns, k)
    if PrefixString is None:
        return "there is no string spelled by the gapped patterns"
    
    SuffixString = StringSpelledByPatterns(SecondPatterns, k)
    if SuffixString is None:
        return "there is no string spelled by the gapped patterns"

    if len(PrefixString) != n + k - 1 or len(SuffixString) != n + k - 1:
        return "there is no string spelled by the gapped patterns"

    for i in range(k + d, len(PrefixString)):
        if PrefixString[i] != SuffixString[i - k - d]:
            return "there is no string spelled by the gapped patterns"

    result = PrefixString + SuffixString[-(k + d):]
    return result

# 主函数示例
if __name__ == "__main__":
    k = 4
    d = 2
    pattern_str = "GAGA|TTGA TCGT|GATG CGTG|ATGT TGGT|TGAG GTGA|TGTT GTGG|GTGA TGAG|GTTG GGTC|GAGA GTCG|AGAT"
    
    gapped_patterns = parse_gapped_patterns(pattern_str)
    if not gapped_patterns:
        print("Error parsing patterns")
    else:
        result = StringSpelledByGappedPatterns(gapped_patterns, k, d)
        print(result)  # 输出: GTGGTCGTGAGATGTTGA
```

```python
from collections import defaultdict

def construct_paired_de_bruijn_graph(paired_kmers):
    graph = defaultdict(list)
    for kmer in paired_kmers:
        left, right = kmer.split("|")
        prefix = (left[:-1], right[:-1])
        suffix = (left[1:], right[1:])
        graph[prefix].append(suffix)
    return graph

def find_eulerian_path(graph):
    in_degrees = defaultdict(int)
    out_degrees = defaultdict(int)

    # 计算入度和出度
    for node, edges in graph.items():
        out_degrees[node] += len(edges)
        for edge in edges:
            in_degrees[edge] += 1

    # 查找起点和终点
    start_node = None
    end_node = None
    all_nodes = set(in_degrees.keys()).union(out_degrees.keys())
    
    # 如果所有节点度数相等，随便选一个作为起点
    for node in all_nodes:
        out_deg = out_degrees.get(node, 0)
        in_deg = in_degrees.get(node, 0)
        
        if out_deg - in_deg == 1:
            start_node = node
        elif in_deg - out_deg == 1:
            end_node = node
        elif out_deg == in_deg:
            continue
        else:
            # 存在度数差不为0或1的节点，无法构成欧拉路径
            raise ValueError("该图不存在欧拉路径")

    # 如果没有起点和终点，说明是欧拉回路
    if not start_node and not end_node:
        start_node = next(iter(all_nodes))

    # Hierholzer算法找欧拉路径
    stack = []
    path = []
    current_node = start_node
    stack.append(current_node)

    while stack:
        if graph[current_node]:
            stack.append(current_node)
            next_node = graph[current_node].pop()
            current_node = next_node
        else:
            path.append(current_node)
            current_node = stack.pop()

    return path[::-1]

def reconstruct_string_from_read_pairs(k, d, paired_kmers):
    graph = construct_paired_de_bruijn_graph(paired_kmers)
    path = find_eulerian_path(graph)

    # 重构前缀和后缀字符串
    prefix_string = path[0][0]
    suffix_string = path[0][1]

    for node in path[1:]:
        prefix_string += node[0][-1]
        suffix_string += node[1][-1]

    # 尝试验证并修复重叠区域
    try:
        # 寻找最大的无冲突重叠区域
        overlap_start = k + d
        max_valid_len = 0
        for i in range(overlap_start, len(prefix_string)):
            if prefix_string[i] == suffix_string[i - overlap_start]:
                max_valid_len = i - overlap_start + 1
            else:
                break

        # 只验证最大有效区域
        if max_valid_len < len(suffix_string) - overlap_start:
            print(f"警告：发现 {len(suffix_string) - overlap_start - max_valid_len} 处碱基不匹配，已使用部分区域进行重构")
        
        # 输出重构结果
        return prefix_string + suffix_string[-(k + d):]
    
    except Exception as e:
        print(f"重构时出现错误: {e}")
        return prefix_string + suffix_string[-(k + d):]

# === 纯字符串手动输入处理 ===
print("请输入配对k-mers字符串（例如：GAGA|TTGA TCGT|GATG）：")
input_str = input().strip()
paired_kmers = input_str.split()

# 自动推断k值
if paired_kmers:
    first_kmer = paired_kmers[0]
    k = len(first_kmer.split("|")[0])
    print(f"自动识别k值为: {k}")

# 手动输入d值
print("请输入d值（两k-mer之间的间隔）：")
d = int(input())

# 执行重构
try:
    result = reconstruct_string_from_read_pairs(k, d, paired_kmers)
    print("\n重构结果：")
    print(result)
except ValueError as ve:
    print(f"输入错误：{ve}")
except Exception as e:
    print(f"运行出错：{e}")
```
