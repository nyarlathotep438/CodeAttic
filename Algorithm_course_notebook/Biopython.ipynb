```python
from Bio.Blast import NCBIWWW, NCBIXML

seq = """TGGGCCTCATATTTATCCTATATACCATGTTCGTATGGTGGCGCGATGTTCTACGTGAATCCACGTTCGAAGGACATCATACCAAAGTCGTAC
AATTAGGACCTCGATATGGTTTTATTCTGTTTATCGTATCGGAGGTTATGTTCTTTTTTGCTCTTTTTCGGGCTTCTTCTCATTCTTCTTTGGCACCTACGGTAGAG"""

# 1. 提交 BLAST（blastn）
result_handle = NCBIWWW.qblast(
    program="blastn",
    database="nt",
    sequence=seq,
    hitlist_size=50,   # 多拿一些结果，方便筛选
)

# 2. 解析 XML
blast_record = NCBIXML.read(result_handle)

# 3. 找到 E 值最低的 hit
best_hit = None
best_e = float("inf")

for alignment in blast_record.alignments:
    for hsp in alignment.hsps:
        if hsp.expect < best_e:
            best_e = hsp.expect
            best_hit = alignment

# 4. 输出结果
if best_hit:
    print("最可能的物种：", best_hit.title)
    print("E-value：", best_e)
else:
    print("没有找到匹配结果")

title = best_hit.title
species = " ".join(title.split()[1:3])  # 取前两个词：Homo sapiens
print("推测物种：", species)
```
